# This notebook reads in processed FAC data and exports a CSV file of all audits for a given state

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
import re
from collections import Counter


# Select state
state = 'HI'


# FAC data
path = '/data/ihwang/data_journalism/allfac'
general_df = pd.read_pickle(path + '/processed/general.pkl')
# Export dataframe
state_df = general_df[general_df['STATE'] == state].copy()
state_df.to_csv(path + '/export/FAC_%s.csv' %state, index=False)
```
